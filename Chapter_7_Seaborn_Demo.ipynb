```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
```

```python
print(iris.shape)
print(iris.info())
iris.head()
```

```python
sns.stripplot(data=iris, x='species', y='petal_length', jitter=True)
plt.show()
```

```python
sns.swarmplot(data=iris, x='species', y='petal_length')
plt.show()
```

```python
sns.set()
```

```python
sns.swarmplot(data=iris, x='species', y='petal_length', palette='Paired')
plt.show()
```

```python
sns.swarmplot(data=iris, x='species', y='petal_length', palette='Reds_r', hue='species')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.boxplot(x='species', y='petal_length',data=iris )
plt.show()
```
